```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
URL = 'https://www.frenchclick.co.uk/t-Brands.aspx'
r = requests.get(URL)
page_body = r.text
```

```python
def find_product(href):
    return href and re.compile(r'm\-.*\.aspx').search(href)
```

```python
soup = BeautifulSoup(page_body, 'html.parser')
all_links = soup.find_all(href = find_product)
```

```python
all_links[0]
```

```python
brand_names = []
for link in all_links:
    brand_names.append(link.contents[0])
```
